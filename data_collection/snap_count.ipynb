```python
import requests
import bs4
import pandas as pd
```

```python
url = 'https://www.footballguys.com/stats/snap-counts/teams?team=PHI&year=2024'
text = requests.get(url).text
```

```python
soup = bs4.BeautifulSoup(text, 'html.parser')
tables = soup.find_all('table')
```

```python
tables
```

```python
theads = tables[0].find_all('thead')
tbodys = tables[0].find_all('tbody')
players = []
for i in range(len(theads)):
    position = theads[i].find('th').text
    test_tr = tbodys[i].find_all('tr')
    for x in test_tr:
        name = ' '.join((x.find('td').text.strip()).split()[0:2])
        snaps = (int(x.find_all('td')[-1].text.strip()))
        players.append({'name': name, 'position': position, 'snaps': snaps})
        
players
    
```

```python
df = pd.DataFrame(players)
df
```

```python
df['snap_pct'] = df['snaps'] / df.groupby('position')['snaps'].transform('sum')
```

```python
df
```

```python
df.to_csv('snap_counts.csv', index=False)
```
